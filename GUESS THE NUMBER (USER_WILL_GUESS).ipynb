```python
'''
Here we have to guess the random number generated by computer whose value is
stored in a variable("random_number")
'''
```

```python
# importing random library to generate random numbers
import random 

# creating function
def guess(x):
    
    # will generate random no. in between from 1 to the no. entered 
    random_number=random.randint(1,x)   #comp. generate a random number
    guess=0
    
    # we are guessing the random_number here
    while (guess!= random_number):
        guess=int(input(f"Enter number between 1 and {x}:"))
        if guess < random_number:
            print("guess number is too low")
        elif guess > random_number:
            print("guess number is too high")

# using f string to concatenate string and int. value            
    print(f"congrats. you have guessed it correct{random_number}") 
guess(10)
```
